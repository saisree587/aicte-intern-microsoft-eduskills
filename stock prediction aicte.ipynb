```python
# --- INSTALLATION SECTION (REQUIRED FOR GOOGLE COLAB) ---
# Run this cell first in your Google Colab notebook to install libraries.
# If you restart your Colab runtime, you will need to run this cell again.

# Install yfinance for fetching stock data
!pip install yfinance

# Optional: Install TA-Lib for advanced technical indicators.
# If you are consistently getting errors with TA-Lib installation,
# you can comment out the entire 'try-except' block below for TA-Lib.
# The script will still run but without the technical indicators.
# NOTE: The TA-Lib installation is being commented out as it frequently
# encounters compilation errors in Colab environments.
# Your script is designed to run without TA-Lib features if it's not installed.
# try:
#     print("Attempting to install TA-Lib C library using apt-get...")
#     # Update package lists
#     !apt-get update -qq > /dev/null
#     # Install TA-Lib C library
#     !apt-get install -y ta-lib -qq > /dev/null
#     print("TA-Lib C library installation complete. Now installing Python binding...")
#     # Install TA-Lib Python binding
#     !pip install ta-lib -qq
#     print("TA-Lib Python binding installation complete.")
# except Exception as e:
#     print(f"Error installing TA-Lib C library or Python binding: {e}")
#     print("Proceeding without TA-Lib. Technical indicators will not be generated.")

# -----------------------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import yfinance as yf
import datetime
import os

# --- IMPORTANT: CRITICAL DISCLAIMER ---
# This code is for educational purposes ONLY and demonstrates a machine learning approach
# to stock price prediction. It does NOT guarantee accuracy, profitability, or reliable
# future stock price predictions. Stock markets are inherently complex, volatile, and
# influenced by innumerable unpredictable factors.
# DO NOT use this code for actual financial investments or decision-making.
# You could lose money. Past performance is not indicative of future results.
# Always consult with a qualified financial advisor before making any investment decisions.
# ----------------------------------------

# Attempt to import talib, if not available, set it to None
try:
    import talib
    print("TA-Lib successfully imported. Technical indicators will be used.")
except ImportError:
    print("TA-Lib is not installed or could not be imported. Technical indicators will not be generated.")
    talib = None # Set to None if not found

def get_technical_indicators(df):
    """
    Calculates various technical indicators for the DataFrame using TA-Lib.
    If TA-Lib is not available, it prints a message and returns the DataFrame unchanged.
    """
    if talib:
        # Simple Moving Average (SMA)
        df['SMA_10'] = talib.SMA(df['Close'], timeperiod=10)
        # Exponential Moving Average (EMA)
        df['EMA_10'] = talib.EMA(df['Close'], timeperiod=10)
        # Relative Strength Index (RSI)
        df['RSI'] = talib.RSI(df['Close'], timeperiod=14)
        # Moving Average Convergence Divergence (MACD)
        macd, macdsignal, macdhist = talib.MACD(df['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
        df['MACD'] = macd
        df['MACD_Signal'] = macdsignal
        # Bollinger Bands (BBANDS)
        upper, middle, lower = talib.BBANDS(df['Close'], timeperiod=20)
        df['BBL_Upper'] = upper
        df['BBL_Middle'] = middle
        df['BBL_Lower'] = lower
        # Average True Range (ATR)
        df['ATR'] = talib.ATR(df['High'], df['Low'], df['Close'], timeperiod=14)
        # On-Balance Volume (OBV)
        df['OBV'] = talib.OBV(df['Close'], df['Volume'])
        print("Technical indicators calculated using TA-Lib.")
    else:
        print("TA-Lib not found. Skipping technical indicator creation.")
    return df

def process_and_predict_stock(ticker, start_date, end_date, prediction_horizon=1,
                              test_size_ratio=0.2, n_estimators=100, random_state=42):
    """
    Processes historical stock data, engineers features, trains a Random Forest Regressor,
    evaluates the model, and predicts the next day's closing price for a given ticker.

    Args:
        ticker (str): The stock ticker symbol (e.g., 'AAPL', 'MSFT').
        start_date (str): Start date for data fetching in 'YYYY-MM-DD' format.
        end_date (str): End date for data fetching in 'YYYY-MM-DD' format.
        prediction_horizon (int): Number of days into the future to predict (default: 1).
        test_size_ratio (float): Proportion of the dataset to include in the test split (default: 0.2).
        n_estimators (int): Number of trees in the Random Forest (default: 100).
        random_state (int): Seed for reproducibility (default: 42).

    Returns:
        dict: A dictionary containing prediction results (RMSE, R2, predicted price, predicted date)
              or None if an error occurs or data is insufficient.
    """
    print(f"\n{'='*50}\n--- Processing Stock: {ticker} ---\n{'='*50}")

    print(f"Fetching data for {ticker} from {start_date} to {end_date}")
    try:
        # Download historical data from Yahoo Finance
        df = yf.download(ticker, start=start_date, end=end_date, progress=False)
        if df.empty:
            print(f"Error: No data fetched for {ticker}. Skipping.")
            return None
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}. Skipping.")
        return None

    print(f"Data downloaded successfully. Total {len(df)} data points.")

    # 1. Feature Engineering
    # Create the target variable: the Close price shifted forward by prediction_horizon days
    df['Target'] = df['Close'].shift(-prediction_horizon)

    # Create lag features (previous day's values)
    df['Close_lag1'] = df['Close'].shift(1)
    df['Volume_lag1'] = df['Volume'].shift(1)
    df['Open_lag1'] = df['Open'].shift(1)
    df['High_lag1'] = df['High'].shift(1)
    df['Low_lag1'] = df['Low'].shift(1)

    # Add technical indicators
    df = get_technical_indicators(df)

    # Drop rows with NaN values. This is crucial as shifting and indicator calculations
    # introduce NaNs at the beginning and end of the DataFrame.
    initial_rows = len(df)
    df.dropna(inplace=True)
    if len(df) < initial_rows:
        print(f"Dropped {initial_rows - len(df)} rows due to NaN values after feature engineering.")

    if df.empty:
        print(f"After dropping NaNs, the dataframe for {ticker} is empty or has insufficient data. Skipping.")
        return None

    # Define features (X) and target (y)
    # Exclude 'Target' (the future value), 'Adj Close' (often redundant with Close),
    # and 'Dividends'/'Stock Splits' (not typically used as direct predictors for price).
    features = [col for col in df.columns if col not in ['Target', 'Adj Close', 'Dividends', 'Stock Splits']]
    X = df[features]
    y = df['Target']

    # Ensure X and y have enough samples after dropping NaNs
    if len(X) <= 1 or len(y) <= 1:
        print(f"Not enough data for {ticker} after feature engineering and dropping NaNs to proceed with splitting. Skipping.")
        return None

    # 2. Split Data into Training and Testing Sets (Time-based split)
    # It's important to split time-series data chronologically to avoid data leakage.
    split_index = int(len(df) * (1 - test_size_ratio))
    if split_index == 0: # Ensure there's at least one training sample
        split_index = 1
    # Check if there's enough data for both training and testing
    if split_index >= len(df):
        print(f"Not enough data for {ticker} to create a test set with ratio {test_size_ratio}. Skipping.")
        return None

    X_train, X_test = X.iloc[:split_index], X.iloc[split_index:]
    y_train, y_test = y.iloc[:split_index], y.iloc[split_index:]
    test_dates = df.index[split_index:] # Keep track of dates for plotting

    if X_test.empty or y_test.empty:
        print(f"Test set for {ticker} is empty after time-based split. Skipping.")
        return None

    print(f"Data split: Training samples = {len(X_train)}, Test samples = {len(X_test)}")

    # 3. Feature Scaling
    # Scale features to prevent features with larger values from dominating the model.
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Convert scaled arrays back to DataFrames, preserving column names and index
    X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
    X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

    print(f"Training Random Forest Regressor for {ticker} with {n_estimators} trees...")
    # 4. Train the Model
    # RandomForestRegressor is an ensemble learning method suitable for regression tasks.
    # n_jobs=-1 uses all available CPU cores for faster training.
    model = RandomForestRegressor(n_estimators=n_estimators, random_state=random_state, n_jobs=-1)
    model.fit(X_train_scaled_df, y_train)
    print("Model training complete.")

    # 5. Make Predictions on the Test Set
    y_pred = model.predict(X_test_scaled_df)

    # 6. Evaluate the Model
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"\n--- Model Evaluation for {ticker} ---")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
    print(f"R-squared (R2): {r2:.2f}")
    print(f"(R2 close to 1 indicates a better fit; for stock prices, R2 is often low due to market volatility)")

    # 7. Visualize the Results
    plt.figure(figsize=(14, 7))
    plt.plot(test_dates, y_test, label='Actual Close Price (Test)', color='blue', linewidth=2)
    plt.plot(test_dates, y_pred, label='Predicted Close Price (Test)', color='red', linestyle='--', linewidth=2)
    plt.title(f'{ticker} Stock Price Prediction (Random Forest - {prediction_horizon}-day horizon)')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.grid(True)
    plt.show()

    # --- Predict the next trading day's price ---
    print(f"\n--- Predicting the next {prediction_horizon} trading day(s) for {ticker} ---")
    # Get the last row of the original DataFrame (before target shift) to use as input for future prediction.
    # This row contains the most recent known features.
    last_known_data_row = df.iloc[-1][features].to_frame().T

    # Scale this single data row using the SAME scaler fitted on the training data.
    last_known_data_scaled = scaler.transform(last_known_data_row)
    last_known_data_scaled_df = pd.DataFrame(last_known_data_scaled, columns=X_train.columns)

    # Make the prediction for the next period
    next_prediction_value = model.predict(last_known_data_scaled_df)[0]

    # Calculate the predicted date (simple increment, might need business day logic for accuracy)
    last_actual_date = df.index[-1]
    predicted_date = last_actual_date + pd.Timedelta(days=prediction_horizon)
    # Simple check to skip weekends (not robust for all market holidays)
    while predicted_date.weekday() > 4: # Monday is 0, Friday is 4
        predicted_date += pd.Timedelta(days=1)

    print(f"Based on data up to {last_actual_date.strftime('%Y-%m-%d')}:")
    # Corrected: Access the scalar value from the Series before formatting
    print(f"Predicted Close Price for approx. {predicted_date.strftime('%Y-%m-%d')}: {next_prediction_value:.2f}")
    print(f"Last actual known Close price: {df['Close'].iloc[-1].item():.2f}") # Corrected line
    print("\n--- REMINDER: This prediction is highly speculative and should NOT be used for financial decisions. ---")

    return {
        'ticker': ticker,
        'rmse': rmse,
        'r2': r2,
        'next_predicted_price': next_prediction_value,
        'predicted_date': predicted_date.strftime('%Y-%m-%d')
    }

if __name__ == "__main__":
    # Define the list of stock tickers you want to predict
    # You can customize this list with any valid stock symbols.
    stock_tickers = ['AAPL', 'MSFT', 'GOOG', 'AMZN', 'TSLA', 'NVDA', 'JPM', 'V', 'PG']

    # Set common parameters for all predictions
    current_date = datetime.date.today()
    end_date = current_date.strftime('%Y-%m-%d')
    # Fetch data for the last 5 years
    start_date = (current_date - datetime.timedelta(days=5*365)).strftime('%Y-%m-%d')
    prediction_days_ahead = 1 # Predict 1 day into the future (next trading day)
    num_trees = 200 # Number of trees for the Random Forest Regressor

    all_prediction_results = []
    for ticker in stock_tickers:
        # Process each stock ticker
        result = process_and_predict_stock(
            ticker,
            start_date,
            end_date,
            prediction_horizon=prediction_days_ahead,
            n_estimators=num_trees
        )
        if result:
            all_prediction_results.append(result)

    print(f"\n\n{'='*60}\n--- Summary of Predictions for All Stocks ---")
    if not all_prediction_results:
        print("No successful predictions were made for any stock.")
    else:
        for res in all_prediction_results:
            print(f"\nTicker: {res['ticker']}")
            print(f"  Test RMSE: {res['rmse']:.2f}")
            print(f"  Test R-squared: {res['r2']:.2f}")
            print(f"  Predicted Price for {res['predicted_date']}: {res['next_predicted_price']:.2f}")
    print(f"{'='*60}")
```
